```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype='int')
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
grid_param={'criterion':['squared_error','absolute_error','poisson','friedman_mse'],'splitter':['best','random'],
            'max_features':['sqrt','log2']}
grid=GridSearchCV(DecisionTreeRegressor(),grid_param,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
result=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(result)
```

```python
table
```

```python
print("The best model in DecisionTree Regressor is {}:".format(grid.best_params_))
```

```python
age=int(input("Enter your age:"))
bmi=float(input("Enter the BMI Value:"))
children=int(input("Enter the number of children:"))
sex=int(input("Enter the sex:"))
smoker=int(input("Are you a Smoker:"))
```

```python
output=grid.predict([[age,bmi,children,sex,smoker]])
output
```
